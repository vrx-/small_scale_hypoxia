```python
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from matplotlib.patches import Rectangle

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

%matplotlib inline
```

```python
ds_xl = xr.open_dataset('files/mean_xl_ts.nc')
ds_l = xr.open_dataset('files/mean_l_ts.nc')
ds_m = xr.open_dataset('files/mean_m_ts.nc')
ds_s = xr.open_dataset('files/mean_s_ts.nc')
ds_xs = xr.open_dataset('files/mean_xs_ts.nc')
ds_2c = xr.open_dataset('files/mean_2c_ts.nc')
```

```python
from scipy import stats
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(4,4), dpi=100)

for ds in [ds_xl, ds_l, ds_m, ds_s, ds_xs, ds_2c]:
    slope, intercept, r_value, p_value, stderr = stats.linregress(ds.hadv, ds.vadv)
    xlr = np.linspace(ds.hadv.min(), ds.hadv.max(), 100)
    ax.scatter(ds.hadv, ds.vadv) 
    ax.plot(xlr, slope*xlr+intercept, 'b--')
    # ax.text(.4, .85, r'$r=%.2f, p=%.3f$' % (r_value, p_value), 
    #         transform=ax.transAxes)

    print('$r=%.2f, p=%.3f$' % (r_value, p_value))
    ax.set_xlabel(r'Horizontal advection [DO kg m$^{-2}$ s$^{-1}$]')
    ax.set_ylabel(r'Vertical advection [DO kg m$^{-2}$ s$^{-1}$]')
# ax.set_xlim(xlr[0], xlr[-1])
```

```python
sns.set_style('whitegrid')
colors = ["black", "ocean blue", "coral", "amber", "crimson"]
colors = ['#000000','#7b3294','#2c7bb6', '#fdae61','#d7191c',]
# sns.set_palette(sns.xkcd_palette(colors))
sns.set_palette(sns.color_palette(colors))
datasets = [ds_xl, ds_m, ds_s, ds_xs]
scales = [r'20x10$^{3}$ km$^{2}$', r'5x10$^{3}$ km$^{2}$',  
          r'500 km$^{2}$', r'50 km$^{2}$', ]


fig, axes = plt.subplots(nrows=4, ncols=1, sharey=True, sharex=True, figsize=(8.5,10), dpi=100)

legend='brief'
for ax, ds, s in zip(axes.flatten(), datasets, scales):
    for coord in ds.coords:
        if coord!='ocean_time':
            ds=ds.drop(coord)
    df = ds.to_dataframe()
    df.columns=[r'$\frac{\partial Ox}{\partial t}$', 
          r'$\vec{v}_h\cdot \nabla_h Ox$',
          r'$w \, Ox$',
          r'$Ak_T \, \frac{\partial{Ox}}{\partial{z}}$', 
          'SOD']
    if s!=scales[0]:
        legend=False
    sns.lineplot(data=df, lw=1, ax=ax, legend=legend)
#     ax.tick_params(labelbottom=False, labelleft=False)
    ax.set_ylim(-2e-2,2e-2)
    ax.set_title(s, pad=-20)

h, l = axes[0].get_legend_handles_labels()
h.insert(1, axes[0].plot(np.NaN, np.NaN, c='w')[0])
l.insert(1,'')
axes[0].get_legend().remove()
axes[0].legend(h,l, ncol=3, frameon=False, loc='lower left', bbox_to_anchor=(0.05, -.05))

axes[0].tick_params(labelleft=True, labelsize = 12)
axes[0].set_ylabel('Domain mean flux \n' r'[DO kg m$^{-2}$ s$^{-1}$]', size=12, color='k')
axes[-1].tick_params(labelbottom=True, labelsize = 12)
axes[-1].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
axes[-1].xaxis.set_minor_locator(mdates.DayLocator(interval=1))
axes[-1].set_xlabel('')
# axes[1,1].set_xlabel('')
axes[-1].xaxis.set_major_formatter(fmt)
axes[-1].set_xlim(ds.ocean_time[0].data,ds.ocean_time[-1].data)

axes[0].ticklabel_format(axis='y', style='sci', scilimits=(0,0))
fig.tight_layout(pad=.3)
fig.savefig('figures/mean_multi.png')
```

```python
fmt = mdates.DateFormatter('%m/%d')
def plot_flux(ax, ds):
        ax.axhline(c='k', ls=':')
        sns.lineplot(ds.ocean_time, ds['rate'], label='rate')
        sns.lineplot(ds.ocean_time, ds['hadv'], label='hadv',)
        sns.lineplot(ds.ocean_time, ds['vadv'], label='vadv',)
        sns.lineplot(ds.ocean_time, ds['vdiff'], label='vdif')
        sns.lineplot(ds.ocean_time, ds['SOD'], label='sod',)
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
        ax.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
        ax.xaxis.set_major_formatter(fmt)
        ax.set_xlim(ds.ocean_time[0].data,ds.ocean_time[-1].data)
```

```python
datasets = [ds_xl, ds_m, ds_s, ds_xs]
scales = [r'20x10$^{3}$ km$^{2}$', r'5x10$^{3}$ km$^{2}$',  
          r'500 km$^{2}$', r'50 km$^{2}$', ]
fig, axes = plt.subplots(nrows=2, ncols=2, sharey=True, sharex=True, figsize=(8.5,6), dpi=100)

for ax, ds, s in zip(axes.flatten(), datasets, scales):
    try:
        ds = ds.drop('s_rho').drop('xi_rho')
    plot_flux(ax, ds)
    ax.tick_params(labelbottom=False, labelleft=False)
    ax.set_ylim(-2e-2,2e-2)
    ax.set_title(s, pad=-20)

axes[0,0].tick_params(labelleft=True, labelsize = 12)
axes[0,0].set_ylabel('VC mean flux \n' r'[DO kg m$^{-2}$ s$^{-1}$]', size=12, color='k')
axes[1,0].tick_params(labelbottom=True, labelsize = 12)
axes[1,0].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
axes[1,0].xaxis.set_minor_locator(mdates.DayLocator(interval=1))
axes[1,0].xaxis.set_major_formatter(fmt)
axes[1,0].set_xlim(ds.ocean_time[0].data,ds.ocean_time[-1].data)
axes[0,0].legend(frameon=False, loc='lower left')
axes[0,0].ticklabel_format(axis='y', style='sci', scilimits=(0,0))
fig.tight_layout(pad=.3)
fig.savefig('figures/mean_multi.png')
```

```python
datasets = [ds_xl, ds_xs,]
scales = [r'20x10$^{3}$ km$^{2}$', r'50 km$^{2}$', ]
filters = [24, 3*24, 7*24]
# labels = ['rate', 'hadv', 'vadv', 'vdiff', 'SOD',]
fig, axes = plt.subplots(nrows=3, ncols=2, sharey=True, sharex=True, figsize=(8.5,8), dpi=100)

for c, ds in enumerate(datasets):
    for r, f in enumerate(filters):
            plot_flux(axes[r,c], ds.rolling(ocean_time=f, center=True).mean())
            axes[r,c].tick_params(labelbottom=False, labelleft=False)
            axes[r,c].set_ylim(-1.2e-2,1.2e-2)
            if r ==0:
                axes[r,c].set_title(scales[c])
axes[0,0].tick_params(labelleft=True, labelsize = 12)
axes[0,0].set_ylabel('VC mean flux \n' r'[DO kg m$^{-2}$ s$^{-1}$]', size=12, color='k')
axes[0,0].ticklabel_format(axis='y', style='sci', scilimits=(0,0))
axes[2,0].tick_params(labelbottom=True, labelsize = 12)
axes[2,0].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
axes[2,0].xaxis.set_minor_locator(mdates.DayLocator(interval=1))
axes[2,0].xaxis.set_major_formatter(fmt)
axes[2,0].set_xlim(ds.ocean_time[0].data,ds.ocean_time[-1].data)
axes[0,0].legend(frameon=False,loc='lower left')
fig.tight_layout(pad=.3)
# fig.savefig('filters.png')

e0_start = mdates.date2num(ds.sel(ocean_time='2010-08-05T06').ocean_time.data)
e0_stop = mdates.date2num(ds.sel(ocean_time='2010-08-07T06').ocean_time.data)
axes[0,1].axvspan(e0_start,  e0_stop, alpha=.2)
axes[0,1].text(e0_start, 1e-2, 'E1', fontsize=8)
e1_start = mdates.date2num(ds.sel(ocean_time='2010-08-10T06').ocean_time.data)
e1_stop = mdates.date2num(ds.sel(ocean_time='2010-08-12T06').ocean_time.data)
axes[0,1].axvspan(e1_start,  e1_stop, alpha=.2)
axes[0,1].text(e1_start, 1e-2, 'E2', fontsize=8)
e2_start = mdates.date2num(ds.sel(ocean_time='2010-09-06T06').ocean_time.data)
e2_stop = mdates.date2num(ds.sel(ocean_time='2010-09-08T06').ocean_time.data)
axes[0,1].axvspan(e2_start,  e2_stop, alpha=.2)
axes[0,1].text(e2_start, 1e-2, 'E3', fontsize=8)

fig.savefig('figures/filters.png')
```
