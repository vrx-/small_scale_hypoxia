```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
import cmocean as cm
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.geoaxes import GeoAxes
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import xroms
from scipy import stats

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
y = 2010
path = '/d1/shared/TXLA_ROMS/output_20yr_obc/%i/ocean_his_00*.nc' % y
ds = xroms.open_roms_netcdf_dataset(path).sel(ocean_time='%i-08' %y)
ds, grid = xroms.roms_dataset(ds, Vtransform=None)

msk = (ds.lon_rho>-95) * (ds.lon_rho<-91) * (ds.h>=10.) * (ds.h<=50.)
msk = msk.load()
msk[165:-1, 440:460]= False
msk[130:160, 300:350]= True
msk[76:83, 490:520]= True
```

```python
def background(ax, extent=[-95, -91, 27.99, 30], p=ccrs.PlateCarree(), res='10m', 
               mers=np.arange(-98, -86), pars=[28, 29, 30],):
    ax.set_extent(extent)
    ax.coastlines(resolution=res)
    ax.add_feature(cfeature.NaturalEarthFeature(category='physical', name='land',
                                                scale=res, facecolor='slategrey'))
    ax.add_feature(cfeature.NaturalEarthFeature(category='physical', name='rivers_lake_centerlines',
                                                scale=res, edgecolor='blue', facecolor='none'))
    if ('PlateCarree' or 'Mercator') in str(p):
        lab = True
    else:
        lab = False
    gl = ax.gridlines(xlocs=mers, ylocs=pars, draw_labels=lab,
                      linewidth=0.5, color='k', linestyle=':')
    gl.xlabels_top = gl.ylabels_right = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    gl.ylabel_style = {'rotation':'90'}
    return gl

```

```python
rho = xroms.density(ds.temp, ds.salt, ds.z_rho)
drhodz = grid.derivative(rho, 'Z', boundary='extend')
```

```python
fig = plt.figure(figsize=(8.5,5), dpi=100)
axes_class = (GeoAxes, dict(map_projection=ccrs.PlateCarree()))
axgr = AxesGrid(fig, 111, axes_class=axes_class,
                nrows_ncols=(2, 2),
                axes_pad=0.1,
                add_all=True,
                share_all=True,
                cbar_mode='None',
                label_mode='')
for i, ax in enumerate(axgr):
    ax.set_aspect(1.13)
    gl = background(ax, extent=[-95, -91, 27.99, 30.01])
    if i!=2:
        gl.xlabels_bottom = gl.ylabels_left = False
        
mp = axgr[0].pcolormesh(H_days.lon_rho, H_days.lat_rho, H_days.where(msk), 
                   cmap='YlOrRd', vmin=0., vmax=31.)
axgr[0].text(.36, .025, 'Exposure [days]',
             horizontalalignment='right', verticalalignment='bottom', color='k',
             transform=axgr[0].transAxes, bbox=dict(facecolor='w', edgecolor='none', pad=0.) )
axgr[0].text(.36, .24, '[%]',
             horizontalalignment='right', verticalalignment='bottom',
             transform=axgr[0].transAxes )
cax = fig.add_axes([.28, 0.56, .2, .03])
cb = fig.colorbar(mp, cax=cax, orientation='horizontal')
cax2 = cax.twiny()
cax2.set_xlim(0,100)
cax2.set_xticks(np.linspace(0,100,5))

mp = axgr[1].pcolormesh(ds.lon_rho, ds.lat_rho, ds.h.where(msk), 
                   cmap=cm.cm.deep,)
axgr[1].text(.36, .025, r'Depth [m]',
             horizontalalignment='right', verticalalignment='bottom', color='k',
             transform=axgr[1].transAxes, bbox=dict(facecolor='w', edgecolor='none', pad=0.) )

cax = fig.add_axes([.67, 0.56, .2, .03])
cb = fig.colorbar(mp, cax=cax, orientation='horizontal',)

mp = axgr[2].pcolormesh(N.lon_rho, N.lat_rho, N, cmap=cm.cm.tempo, vmin=20, vmax=40)
axgr[2].text(.36, .025, r'N [cph]',
             horizontalalignment='right', verticalalignment='bottom', color='k',
             transform=axgr[2].transAxes, bbox=dict(facecolor='w', edgecolor='none', pad=0.) )
cax = fig.add_axes([.28, 0.17, .2, .03])
cb = fig.colorbar(mp, cax=cax, orientation='horizontal',extend='both')

mp = axgr[3].pcolormesh(bbl_m.lon_rho, bbl_m.lat_rho, bbl_m, 
                   cmap='bone_r', vmin=0, vmax=6.)
t = axgr[3].text(.36, .025, r'$\overline{BBL}_{thk}$ [m]',
             horizontalalignment='right', verticalalignment='bottom', color='k',
             transform=axgr[3].transAxes, bbox=dict(facecolor='w', edgecolor='none', pad=0.))
cax = fig.add_axes([.67, 0.17, .2, .03])
cb = fig.colorbar(mp, cax=cax, orientation='horizontal', extend='max')

for ax, l in zip(axgr, ['(a)', '(b)', '(c)', '(d)']):
    ax.text(-91.5, 29.7, l, fontsize=13)

fig.savefig('figures/ocurrence.png')
```

```python
budget = xr.open_dataset('files/budget.nc')
sod_08 = budget.SOD.sel(ocean_time='2010-08').where(msk, drop=True)
sod_m = sod_08.mean('ocean_time').compute()
```

```python
# BBL_ = xr.open_dataarray('files/full_bbl2_2010_08.nc').where(msk, drop=True).sel(ocean_time='2010-08')                                               
# H_days = xr.open_dataarray('files/H_days_2010_08.nc').where(msk, drop=True)
# N2_mean = xr.open_dataarray('files/N2_mean_2010_08.nc').where(msk, drop=True)
# N = np.sqrt(N2_mean)*(3600*.5/np.pi)
nbins=60

fig, axes = plt.subplots(2, 2, figsize=(8 ,8), dpi=100, sharex='col', sharey='row')

#1
bbl = bbl_m.values.flatten()
thyp = H_days.where(msk, drop=True).values.flatten()
sod = sod_m.values.flatten()
bf = N.values.flatten()

xymsk = np.isfinite(thyp)*np.isfinite(bbl)*np.isfinite(sod)*np.isfinite(bf)

bbl = bbl[xymsk]
thyp = 100/31*thyp[xymsk]

xlr = np.asarray([bbl.min(),bbl.max()])
slope, intercept, r_value, p_value, stderr = stats.linregress(bbl, thyp)


cbbl = axes[0,0].hist2d(bbl, thyp, bins=nbins, cmap='binary', vmin=0)
axes[0,0].plot(xlr, slope*xlr+intercept, 'b--')
axes[0,0].text(.7, .65, 'r=%.2f' % r_value, 
        transform=axes[0,0].transAxes, color='b')

axes[0,0].set_xlim(xlr[0], xlr[-1])
axes[0,0].set_ylim(0,100)
axes[0,0].set_ylabel(r'Hypoxic exposure [%]')
print(p_value)

#2
bf = bf[xymsk]

xlr = np.asarray([bf.min(),bf.max()])
slope, intercept, r_value, p_value, stderr = stats.linregress(bf, thyp)

cN = axes[0,1].hist2d(bf, thyp, bins=nbins, cmap='binary', vmin=0)
axes[0,1].plot(xlr, slope*xlr+intercept, 'b--')
axes[0,1].text(.7, .65, 'r=%.2f' % r_value, 
        transform=axes[0,1].transAxes, color='b')
axes[0,1].set_xlim(xlr[0], xlr[-1])
print(p_value)

#3
sod = -sod[xymsk]

xlr = np.asarray([bbl.min(),bbl.max()])
slope, intercept, r_value, p_value, stderr = stats.linregress(bbl, sod)

cbbl = axes[1,0].hist2d(bbl, sod, bins=nbins, cmap='binary', vmin=0)
axes[1,0].plot(xlr, slope*xlr+intercept, 'b--')
axes[1,0].text(.7, .65, 'r=%.2f' % r_value, 
        transform=axes[1,0].transAxes, color='b')

axes[1,0].set_ylim(0, 6e-4)
axes[1,0].set_xlabel(r'BBL thickness [m]')
axes[1,0].set_ylabel(r'SOD [mmol m$^{-2}$ s$^{-1}$]')
axes[1,0].ticklabel_format(style='sci', axis='y', scilimits=(0,0))
print(p_value)
#4
xlr = np.asarray([bf.min(),bf.max()])
slope, intercept, r_value, p_value, stderr = stats.linregress(bf, sod)

cN = axes[1,1].hist2d(bf, sod, bins=nbins, cmap='binary', vmin=0)
axes[1,1].plot(xlr, slope*xlr+intercept, 'b--')
axes[1,1].text(.7, .65, 'r=%.2f' % r_value, 
        transform=axes[1,1].transAxes, color='b')

axes[1,1].set_xlabel(r'Buoyancy frequency [cph]')
axes[1,1].set_xlim(xlr[0], xlr[-1])
axes[1,1].set_ylim(0, -6e-4)
print(p_value)
fig.savefig('figures/correlation.png')
```
